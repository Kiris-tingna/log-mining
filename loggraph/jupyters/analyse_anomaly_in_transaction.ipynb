```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.ticker as ticker

file1 = '../data/BasicSignatureGrenGini.csv'
file2 = '../../../event_type_ansible.csv'
df1 = pd.read_csv(file1, nrows=60000, index_col='time')
df2 = pd.read_csv(file2, nrows=60000, index_col='time')
df1.index = pd.to_datetime(df1.index,  format="%Y-%m-%d %H:%M:%S.%f")
df2.index = pd.to_datetime(df2.index,  format="%Y-%m-%d %H:%M:%S.%f")
data1 = df1.sort_index()
data2 = df2.sort_index()
```

```python
dataframe1 = data1[data1['level']=='INFO']
print(len(dataframe1))
dataframe1.sample(5)
dataframe2 = data2[data2['level']=='INFO']
print(len(dataframe2))
dataframe2.sample(5)
```

```python
first_deploy_1 = dataframe1[(dataframe1['deploy_id']==59442) & ((dataframe1['component']== 'cf-pdman') | (dataframe1['component']== 'pdm-cli'))]
second_deploy_1 = dataframe1[(dataframe1['deploy_id']==59445) & ((dataframe1['component']== 'cf-pdman') | (dataframe1['component']== 'pdm-cli'))]

N1 = len(first_deploy_1)
ind1 = np.arange(N1)

def format_date_1(x, pos=None):
    thisind = np.clip(int(x+0.5), 0, N1-1)
    return first_deploy_1.index[thisind].strftime('%Y-%m-%d %H:%M:%S')

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,1,1)
ax1.scatter(ind1, first_deploy_1['event_id'])
ax1.xaxis.set_major_formatter(ticker.FuncFormatter(format_date_1))
fig.autofmt_xdate()


N2 = len(second_deploy_1)
ind2 = np.arange(N2)

def format_date_2(x, pos=None):
    thisind = np.clip(int(x+0.5), 0, N2-1)
    return second_deploy_1.index[thisind].strftime('%Y-%m-%d %H:%M:%S')

fig = plt.figure(figsize=(15,5))
ax2 = fig.add_subplot(1,1,1)
ax2.scatter(ind2, second_deploy_1['event_id'])
ax2.xaxis.set_major_formatter(ticker.FuncFormatter(format_date_2))
fig.autofmt_xdate()
```
